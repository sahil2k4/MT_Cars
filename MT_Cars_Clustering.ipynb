```python
df=df.rename(columns={'Unnamed: 0': 'Model'})
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
df['Model']=label.fit_transform(df.Model)
```

```python
print(df.info())
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.drop('Model', axis=1))
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia = []
k_range = range(1, 15)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title("Elbow Method For Optimal k")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
# Let's say the elbow is at k=4
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, df['Cluster'])
print(f" Score: {score:.2f}")
```

```python
from sklearn.decomposition import PCA
import seaborn as sns

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df['PCA1'], df['PCA2'] = X_pca[:, 0], X_pca[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PCA1', y='PCA2', hue='Cluster', palette='Set2')
plt.title("K-Means Clusters (PCA Reduced)")
plt.show()
```

```python
cluster_summary = df.groupby('Cluster').mean()
print(cluster_summary)
```

```python
sns.pairplot(df, hue='Cluster', diag_kind='kde')
```

1. Cluster Separation is Clear in Some Features
mpg (Miles Per Gallon) and wt (Weight) show strong inverse patterns:

One cluster tends to have high mpg and low weight (likely economy cars).

Another cluster has low mpg and high weight (likely performance or luxury cars).

2. Cluster Structure Appears in Power & Engine Features
Variables like hp (Horsepower), qsec (Quarter mile time), and disp (Displacement) show some cluster separation:

One cluster has high hp and disp, probably heavy-duty or sports cars.

Another cluster has lower hp and faster qsec, indicating quick, lighter cars.

3. Cylinders (cyl), Gears (gear), and Carburetors (carb) show step-like, categorical grouping
These features create horizontal and vertical striping, showing:

Clusters align partially with the number of cylinders (e.g., 4, 6, 8).

One cluster is mostly 4-cylinder cars, another mostly 8-cylinder.

4. Some Features Offer Little Separation
Features like drat (rear axle ratio) or vs (engine shape) show minimal cluster separation.

These may not contribute much to the clustering result and could be dropped for simplification.
